## Word2Vec 사용

```python
from gensim.models import Word2Vec 
```

```python
# 모델 로딩
pre_model = Word2Vec.load('pre_trained/word2vec')
my_model = Word2Vec.load('wiki_w2v')
```

```python
my_model.wv.save_word2vec_format('my_w2v_save')
```

```python
pre_model.wv.save_word2vec_format('pre_w2v_save')
```

```python
# 학습된 결과 비교
def most_sim(pos=[], neg=[], n=10, pre=pre_model, mine=my_model):
    print('Pre-trained Embeddings:\n', pre.wv.most_similar(positive=pos, negative=neg, topn=n))
    print('My Embeddings:\n', mine.wv.most_similar(positive=pos, negative=neg, topn=n))

def most_sim_cosmul(pos=[], neg=[], n=10, pre=pre_model, mine=my_model):
    print('Pre-trained Embeddings:\n', pre.wv.most_similar_cosmul(positive=pos, negative=neg, topn=n))
    print('My Embeddings:\n', mine.wv.most_similar_cosmul(positive=pos, negative=neg, topn=n))
```

```python
def most_sim_word(pos=[], neg=[], n=10, pre=pre_model):
    data = pre.wv.most_similar(positive=pos, negative=neg, topn=n)
    wordlist = []
    for word, vector in data:
        wordlist.append(word)
    return wordlist
```

```python
most_sim(pos=['서울시', '청년', '지원', '센터'], neg=[], n=30)
```

```python
most_sim(pos=['서울시', '관광', '여행'], neg=[], n=30)
```

```python
most_sim_word(pos=['서울시', '청년', '지원', '센터'], n=10)
```

```python
most_sim_word(pos=['서울시', '관광', '여행'], n=20)
```

```python
word = "사과"

print("'{}'의 유사도 검색결과:\n".format(word), most_sim_word(pos=[word, '포도', '복숭아'], n=10))
```

```python
word = "머리"
print("'{}'의 유사도 검색결과:\n".format(word), most_sim_word(pos=[word], n=10))
print("'{}'의 유사도 검색결과:\n".format(word), most_sim_word(pos=[word, '미용'], n=10))
```
